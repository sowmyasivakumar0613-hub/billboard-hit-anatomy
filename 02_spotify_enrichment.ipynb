```python
import pandas as pd

# Your original top-per-year data
top = pd.read_csv(r"C:\Users\Jiyoon\billboard-project\data\processed\top_per_year.csv")

# The new enriched dataset
enriched = pd.read_csv(r"C:\Users\Jiyoon\billboard-project\data\raw\Hot 100 Audio Features.csv")


print("Top per year shape:", top.shape)
print("Enriched shape:", enriched.shape)
```

```python
# clean top_per_year
top['song_clean'] = top['song'].str.lower().str.strip()
top['artist_clean'] = top['artist'].str.lower().str.strip()

#clean enriched dataset
enriched['song_clean'] = enriched['Song'].str.lower().str.strip()
enriched['artist_clean'] = enriched['Performer'].str.lower().str.strip()

print("Cleaning Done!")
```

```python
merged = top.merge(
    enriched[['song_clean', 'artist_clean', 'danceability', 'energy', 'valence', 'tempo', 'acousticness', 'speechiness',
              'instrumentalness', 'liveness', 'loudness']],
    on = ['song_clean', 'artist_clean'],
    how = 'left'
)

print("Merged shape: ", merged.shape)
print("Rows with audio features: ",merged['energy'].notna().sum())
print("Rows missing audio features:",merged['energy'].isna().sum())
```

```python
# matched songs
matched = merged[merged['energy'].notna()][['year', 'song', 'artist', 'tempo', 'energy', 'valence']]
print("MATCHED SONGS: ")
print(matched)

#didn't match
missing = merged[merged['energy'].isna()][['year','song', 'artist']]
print("\nMISSING SONGS: ")
print(missing)
```

```python
enriched_dedup = enriched.drop_duplicates(subset=['song_clean', 'artist_clean'], keep='first')
print("Before:", enriched.shape[0], "After:", enriched_dedup.shape[0])
```

```python
merged = top.merge(
    enriched_dedup[['song_clean', 'artist_clean', 'danceability', 'energy', 
               'valence', 'tempo', 'acousticness', 'speechiness',
               'instrumentalness', 'liveness', 'loudness']],
    on=['song_clean', 'artist_clean'],
    how='left'
)

print("Merged shape:", merged.shape)  # should now be (64, 16)
```
