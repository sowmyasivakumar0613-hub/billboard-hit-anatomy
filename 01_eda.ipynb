```python
df["date"] = pd.to_datetime(df["date"])
df.dtypes
```

```python
df["year"] = df["date"].dt.year
df["decade"] = (df["year"] // 10) * 10

df[["date" , "year" , "decade"]].head()
```

```python
df["decade"].value_counts().sort_index()
```

```python
number_ones = df[df['rank'] == 1].copy()

# Count how many weeks each song held #1, PER YEAR (not lifetime)
weeks_at_no1_per_year = (
    number_ones
    .groupby(['year', 'decade', 'song', 'artist'])
    .size()
    .reset_index(name='weeks_at_no1_this_year')
)

# Now pick the song with the most #1 weeks within each year
top_per_year = (
    weeks_at_no1_per_year
    .sort_values('weeks_at_no1_this_year', ascending=False)
    .drop_duplicates('year', keep='first')
    .sort_values('year')
    [['year', 'decade', 'song', 'artist', 'weeks_at_no1_this_year']]
)

print(top_per_year.shape)
top_per_year[top_per_year['year'] >= 2019]
```
